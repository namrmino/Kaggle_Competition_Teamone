# train과 test 데이터셋 Z-test 검정

```
del train['id']
del test['id']
```

```
# 범주형 변수와 수치형 변수 나누기
cat_columns = [c for c, t in zip(train.dtypes.index, train.dtypes) if t == 'O'] 
num_columns = [c for c in train.dtypes.index if c not in cat_columns]

print('범주형 변수: \n{}\n\n 수치형 변수: \n{}\n'.format(cat_columns, num_columns))
```

# 데이터 전처리

```
tmp_train = train.copy()
tmp_test = test.copy()
```

```
# 제거시키기로한 변수 drop
tmp_train = tmp_train.drop(['workclass', 'education', 'relationship', 'native_country'], axis=1)
tmp_test = tmp_test.drop(['workclass', 'education', 'relationship', 'native_country'], axis=1)
```

```
tmp_train.head()
```

```
tmp_test.head()
```

```
# marital_status 이원화
tmp_train.marital_status.value_counts()
```

```
tmp_train['marital_status'] = tmp_train['marital_status'].apply(lambda x: 1 if x =='Married-civ-spouse' else 0)
tmp_train['marital_status'].value_counts()
```

```
tmp_train.head()
```

```
# ? => other 카테고리로 바꿔주기
has_na_columns = ['occupation']
for c in has_na_columns:
  tmp_train.loc[train[c] == '?', c] = 'other'
  tmp_test.loc[test[c] == '?', c] = 'other'
(tmp_train[has_na_columns] == '?').sum()
```

```
tmp_train.sample(5)
```

```
# 범주형 변수 인코딩

from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse=False)
ohe_columns = ohe.fit(tmp_train[['occupation', 'race', 'sex']])
```

```
ohe_columns = list()
for lst in ohe.categories_:
    ohe_columns += lst.tolist()
```

```
new_train_cat = pd.DataFrame(ohe.transform(tmp_train[['occupation', 'race', 'sex']]), columns=ohe_columns)
```

```
new_train_cat.head()
```

```
# 인코딩데이터와 기존tmp_train데이터 합치기
tmp_train = pd.concat([tmp_train, new_train_cat], axis=1)

# 기존 범주형 변수 제거
tmp_train = tmp_train.drop(columns=['occupation', 'race', 'sex'])
```

```
tmp_train
```

```
# capital_gain, capital_loss 로그화 전처리
tmp_train['log_capital_gain'] = train['capital_gain'].map(lambda x : np.log(x) if x != 0 else 0)
tmp_test['log_capital_gain'] = test['capital_gain'].map(lambda x : np.log(x) if x != 0 else 0)
tmp_train['log_capital_loss'] = train['capital_loss'].map(lambda x : np.log(x) if x != 0 else 0)
tmp_test['log_capital_loss'] = test['capital_loss'].map(lambda x : np.log(x) if x != 0 else 0)
tmp_train = tmp_train.drop(columns=['capital_loss', 'capital_gain'])
tmp_test  = tmp_test.drop(columns=['capital_loss', 'capital_gain'])
```
