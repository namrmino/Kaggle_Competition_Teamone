```python
import os
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings(action='ignore')
```

```python
train = pd.read_csv('~/Downloads/kakr-4th-competition/train.csv')
test = pd.read_csv('~/Downloads/kakr-4th-competition/test.csv')
```

```python
train.head(10)
```

```python
# 불필요한 변수 삭제 
# 'workclass', 'occupation' 변수 함께 삭제함 - 처리 방법 고려 필요..

train = train.drop(['id', 'education', 'relationship', 'native_country', 'workclass', 'occupation'], axis=1)
train
```

```python
# target 값 분리
target = train['income']
target
```

```python
train = train.drop('income', axis=1)
train
```

```python
# marital_status 정리

train.marital_status.value_counts()
```

```python
train['marital_status']=='Married-civ-spouse'
```

```python
train['marital_status'] = train['marital_status'].apply(lambda x: 1 if x =='Married-civ-spouse' else 0)
train['marital_status'].value_counts() 
```

```python
# 범주형 변수 인코딩

from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse=False)
ohe_columns = ohe.fit(train[['race', 'sex']])
```

```python
ohe_columns = list()
for lst in ohe.categories_:
    ohe_columns += lst.tolist()
```

```python
new_train_cat = pd.DataFrame(ohe.transform(train[['race', 'sex']]), columns=ohe_columns)
```

```python
new_train_cat.head()
```

```python
train = pd.concat([train, new_train_cat], axis=1)


# 기존 범주형 변수 제거
train = train.drop(columns=['race', 'sex'])
```

```python
train
```

```python
## 데이터 분할

from sklearn.model_selection import train_test_split

x_train, x_valid, y_train, y_valid = train_test_split(train, target,
                                                     test_size=.3,
                                                     random_state = 42,
                                                     shuffle=True,
                                                     stratify = target)
```

```python
x_train.columns
```

```python
# Decision Tree 모델 사용

from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier(random_state=2020)

dt_clf.fit(x_train, y_train)
```

```python
# 모델 평가 
from sklearn.metrics import f1_score

y_pred = dt_clf.predict(x_valid)

print(f"DecisionTreeClassifier F1 Score: {f1_score(y_valid, y_pred, average='micro')}")
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_valid, y_pred)
print(np.round(accuracy, 4))
```

```python
# train 학습 결과
y_pred_train = dt_clf.predict(x_train)
y_pred_train = pd.DataFrame(y_pred_train)

train_set_predict = pd.DataFrame(y_train)
train_set_predict['predict'] = y_pred_train.values
train_set_predict.rename(columns={'income':'true'}, inplace=True)
print(train_set_predict.head(10))
```

```python
# 검증 예측 결과

y_pred = dt_clf.predict(x_valid)

vaild_set_predict = pd.DataFrame(y_valid)
vaild_set_predict['predict'] = y_pred
vaild_set_predict.rename(columns={'income':'true'}, inplace=True)
print(vaild_set_predict.head(10))
print('='*50)
print('*전체 검증세트 수: {} \n 실제 예측 수 : {}'.format(vaild_set_predict['true'].count(), 
                            (vaild_set_predict['true'].values==vaild_set_predict['predict'].values).sum()))
```
